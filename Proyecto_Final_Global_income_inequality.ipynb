# **¿Qué podria suceder en el día de mañana?**

En esta etapa, investigaremos según la información de nuestra base de datos (según los intereses que venimos teniendo), que se puede visualizar a un futuro próximo, si hay probabilidades de mejoras o no en el área socioeconómico de cada país.
Es clave destacar que esto puede variar debido a que hay microfactores que influyen a que la economía de un país mejore o empeore. Pero aquí nos basaremos en la información obtuvida de los años 2000 al 2023 para tener un panorama de lo que pueda suceder en el 2024 y 2025.

### **Algoritmo de predicción: LinearRegression.**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# preparación del model de datos
df_prediction = df[['Country', 'Year', 'Gini Index']]
countries = df_prediction['Country'].unique()
predictions = {}
accuracy = {}

for country in countries:
    country_data = df_prediction[df_prediction['Country'] == country]
    X = country_data['Year'].values.reshape(-1, 1)
    y = country_data['Gini Index'].values.reshape(-1, 1)

    # Entrenamiento en el modelo de regresión lineal
    model = LinearRegression()
    model.fit(X, y)

    # Predicción Gini para los futuros 2 años siguientes
    future_years = np.array([max(df_prediction['Year']) + 1, max(df_prediction['Year']) + 2]).reshape(-1, 1)
    predicted_gini = model.predict(future_years)

    predictions[country] = predicted_gini

    # Calculo de certeza en la predicción
    y_pred = model.predict(X)
    rmse = mean_squared_error(y, y_pred, squared=False)
    accuracy[country] = 1-rmse

    # Print
for country, pred in predictions.items():
    print(f"Predictions for {country}:")
    for i, year in enumerate(future_years.flatten()):
        print(f"  Year {int(year)}: {pred[i][0]:.4f}")
    print(f"Model accuracy for {country}: {accuracy[country]:.4f}")
```

```python
from sklearn.linear_model import LinearRegression

# Grupo país y años de la base de datos
grouped_df = df.groupby(['Country', 'Year'])['Gini Index'].mean().reset_index()

# Predicción
predictions = {}

# Países
for country in grouped_df['Country'].unique():
    # filtro
    country_data = grouped_df[grouped_df['Country'] == country]

    # Preparación de linea de regresión
    X = country_data['Year'].values.reshape(-1, 1)
    y = country_data['Gini Index'].values

    # Entrenamiento del modelo
    model = LinearRegression()
    model.fit(X, y)

    # Predicción de gini próximos 2 años
    next_two_years = np.array([country_data['Year'].max() + 1, country_data['Year'].max() + 2]).reshape(-1, 1)
    predicted_gini = model.predict(next_two_years)

    # Predicción
    predictions[country] = predicted_gini

# Predicciones
for country, gini_values in predictions.items():
    print(f"Predictions for {country}:")
    for i, gini in enumerate(gini_values):
        print(f"  Year {int(grouped_df[grouped_df['Country'] == country]['Year'].max()) + i + 1}: {gini:.4f}")
```

Aquí tenemos un primer resultado de predicción, si bien ambos son lo mismo, uno indica la presición del modelo elegido para predecir el porcentaje de Gini de cada país en los años 2024 y 2025.
Es interesanta ver que el resultado de certeza en cada predicción está entre el 0.85 al 0.89%, por el cual es muy positivo incluso cercano a porcentajes reales y posibles.

```python
# Comparación de la predicción de Gini index 2024 y 2025 de cada páis con los registrado del 2022 y 2023 del gini de los países.

import pandas as pd
import numpy as np

prediction_df = pd.DataFrame.from_dict(predictions, orient='index', columns=['2024', '2025'])
prediction_df = prediction_df.reset_index().rename(columns={'index': 'Country'})


gini_2023 = df[df['Year'] == 2023].groupby('Country')['Gini Index'].mean()


comparison_df = prediction_df.merge(gini_2023, on='Country', how='left')
comparison_df.rename(columns={'Gini Index': '2023'}, inplace=True)


def compare_gini(row):
    results = []
    for year in ['2024', '2025']:
        if row[year] < row['2023']:
            results.append('Positive')
        elif row[year] > row['2023']:
          results.append('Negative')
        else:
          results.append('Equal')
    return results


comparison_df[['2024_result', '2025_result']] = comparison_df.apply(compare_gini, axis=1, result_type='expand')



comparison_df
```
